```python
import cv2 as cv

import numpy as np
from matplotlib import pyplot as plt

import seaborn as sns
sns.set_theme()
```

```python
# normalize image
img = cv.cvtColor(cv.imread('highsierra.jpg'), cv.COLOR_BGR2GRAY)
img_norm = cv.normalize(img.astype(float), None, 0, 1, cv.NORM_MINMAX)

plt.axis(False)
plt.imshow(img_norm, cmap='gray')
```

```python
# apply box filter
img_box = cv.blur(img, (10, 10))

plt.axis(False)
plt.imshow(img_box, cmap='gray')
```

```python
# apply Gaussian blur from a kernel with a variance of 10
img_gauss = cv.GaussianBlur(img, (0, 0), 10)

plt.axis(False)
plt.imshow(img_gauss, cmap='gray')
```
